```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 超参数搜索结果 (示例数据)
alpha_values = [10,8,6,4,2]
beta_values = [0.2, 0.25, 0.3, 0.35, 0.4]
accuracy = np.array([
    [87.2, 87.40, 87.59, 87.78, 87.79],
    [88.33, 88.31, 88.36, 88.40, 88.27],
    [88.12, 88.31, 88.19, 88.18, 88.22],
    [87.20, 87.6, 87.14, 87.14, 87.2],
    [86.97, 87.21, 87.16, 86.79, 86.70]
])

# 创建热力图
plt.figure(figsize=(8, 6))
sns.heatmap(accuracy, annot=True, fmt=".2f", xticklabels=beta_values, yticklabels=alpha_values, cmap="YlGnBu")
'''
将 cmap 替换为以下任意一种：

"viridis"（默认配色，适合多维度数据）
"plasma"（鲜艳渐变）
"coolwarm"（蓝红对比，适合显示高低对比）
"YlGnBu"（黄色到绿色到蓝色，适合正值变化）
"cividis"（对比度高，适合色盲友好型）
'''

# 添加标题和标签
plt.title("Hyperparameter Search on CIFAR-10 (Imbalance Ratio: 100)")
plt.xlabel(r"$\mu_1$ (Coefficient of $\mathcal{L}_{\mathrm{sim}}$)")
plt.ylabel(r"$\mu_2$ (Coefficient of $\mathcal{L}_{\mathrm{con}}$)")

# 显示图表
plt.tight_layout()
plt.savefig('./Hyperparameter Search.png',dpi=300,facecolor = 'w')
plt.show()
```
